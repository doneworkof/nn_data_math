```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
a = np.random.randn()
b = np.random.randn()
```

```python
a_true = 2
b_true = -1
```

```python
def f_true(x):
    return a_true * x + b_true
```

```python
def f(x):
    return a * x + b
```

```python
x_train = np.random.randint(low=-10, high=11, size=(15,))
```

```python
y_train = f_true(x_train)
```

```python
def plot_both():
    x_range = np.arange(-50, 51)
    y_true = f_true(x_range)
    y_pred = f(x_range)
    fig, ax = plt.subplots()
    plt.plot(x_range, y_true, color='blue', label='true')
    plt.plot(x_range, y_pred, color='red', label='pred')
    plt.legend()
    ax.axhline(y=0, color='grey')
    ax.axvline(x=0, color='grey')
    ax.set_xlabel('x value')
    ax.set_ylabel('y value')
    plt.grid()
    plt.show()
```

```python
plot_both()
```

```python
epochs = 50
lr = 0.005

for epoch in range(epochs):
    for x, y_true in zip(x_train, y_train):
        y_pred = f(x)
        E = (y_pred - y_true) ** 2
        dE_dsub = 2 * (y_pred - y_true)
        dE_da = dE_dsub * x
        dE_db = dE_dsub
        a -= dE_da * lr
        b -= dE_db * lr
        print('Loss:', E)
        
```
